### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
df_cities["population_total"].sort_values(ascending=False)
filtro = df_cities["population_total"] > 500000
df_cities_top_5 = df_cities[filtro].sort_values(by="population_total",ascending=False)
df_cities_top_5
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(9,9)

colores = ["#7F7FFF","#BFBFBF","#FF7F7F","#BFBF7F","#D19494"] # --> Los he sacado con la herramienta selector de colores de Paint

ax.bar(df_cities_top_5["city"], height = df_cities_top_5["population_total"]/1000, color = colores)
ax.set_ylabel("Habitantes (miles)")
ax.set_ylim(0,5000)
ax.set_title("Ciudades mas pobladas de California");

ax.text(x = 0, y = (df_cities_top_5.iloc[0,6]/1000)+100, s = str(df_cities_top_5.iloc[0,6]/1000), horizontalalignment = "center")
ax.text(x = 1, y = (df_cities_top_5.iloc[1,6]/1000)+100, s = str(df_cities_top_5.iloc[1,6]/1000), horizontalalignment = "center")
ax.text(x = 2, y = (df_cities_top_5.iloc[2,6]/1000)+100, s = str(df_cities_top_5.iloc[2,6]/1000), horizontalalignment = "center")
ax.text(x = 3, y = (df_cities_top_5.iloc[3,6]/1000)+100, s = str(df_cities_top_5.iloc[3,6]/1000), horizontalalignment = "center")
ax.text(x = 4, y = (df_cities_top_5.iloc[4,6]/1000)+100, s = str(df_cities_top_5.iloc[4,6]/1000), horizontalalignment = "center")
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(9,9)


ax.hist(df_cities["area_total_km2"],
         label="Hist Area",
         bins = 30,
         color= "steelblue",
         alpha = 0.8
         )

font1 = {"family":"Bradley Hand itc","size": 18}
ax.set_ylabel("Nmero de Ciudades", fontdict= font1)
ax.set_xlabel("Superficie ($Km^2$)", fontdict= font1)
ax.set_title("Poblacin por km cuadrado de agua")
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(9,9)


ax.scatter(df_cities["area_water_sq_mi"],df_cities["population_total"]/1000, color="green", alpha = 0.5)

coordenadas1 = (34, 3884.4307)
coordenadas2 = (47.21, 1345.895)

ax.annotate("Outliers", xy = coordenadas1,  xycoords='data', xytext=(0.5,0.85), textcoords="axes fraction", arrowprops=dict(facecolor = "black", arrowstyle= "->"))
ax.annotate("Outliers", xy = coordenadas2,  xycoords='data', xytext=(0.5,0.85), textcoords="axes fraction", arrowprops=dict(facecolor = "black", arrowstyle= "->"))


ax.set_ylabel("Poblacin (miles habitantes)")
ax.set_xlabel("rea de agua en la ciudad ($Km^2$)")
plt.show()
```

```python
filtro_agua = df_cities.area_water_sq_mi > 35
filtro_poblacion = df_cities.population_total > 1200000
df_cities.loc[filtro_poblacion | filtro_agua, ["population_total", "area_water_sq_mi"]]
```

```python

df_cities.dropna(subset=["elevation_m"], inplace=True)
df_cities.info()
```

```python

fig, ax = plt.subplots()
fig.set_size_inches(9,9)

ax.boxplot(df_cities["elevation_m"],
           patch_artist = True,
           boxprops = dict(facecolor = "red"),
           medianprops = dict(color = "black"),
           vert = False);

ax.set_xlabel("Altitud (m)")
ax.set_title("Diagrama de caja de la Altitud de las ciudades de California")

plt.show()
```
